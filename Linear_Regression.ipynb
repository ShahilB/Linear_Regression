# Linear_Regresion 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("weight-height.csv")
df.head()
```

```python
plt.scatter(df['Height'],df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= df[['Height']]
y= df[['Weight']]
x.shape , y.shape
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=2000, random_state= 42)
```

```python
x_train.shape, x_test.shape
```

```python
#Standardize the dataset train independent data
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train= scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
plt.scatter(x_train,y_train)
```

```python
## Train the Simple Linear Regression 
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print(f"Slope or coefficient of weight : {model.coef_}" )
print(f"Intercept : {model.intercept_}")
```

```python
# Predict the Y variable
y_pred= model.predict(x_train)
```

```python
plt.scatter(x_train, y_train, color='blue', label='Data points')
plt.plot(x_train, y_pred, color='red', label='Fitted line')
plt.xlabel("Height")
plt.ylabel("Weight")
```

```python
y_pred_test= model.predict(x_test)
```

```python
y_pred_test,y_test
```

## Performance Matrcies 
### MSE MAE RMSE

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse =mean_squared_error(y_test,y_pred_test)
mae = mean_absolute_error(y_test,y_pred_test)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred_test)
print(score)
```
